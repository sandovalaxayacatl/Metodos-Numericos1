$f(x)= \sqrt{x+1}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#valores del eje x que toma en la grafica.
x=np.linspace(-5,5,100)
```

```python
#Funcion cuadratica.
def f(x):
    return (x+1)**(1/2)
def g(x):
    return (1+(1/2)*x)
def h(x):
    return (1((1/2)*x)-(1/8)*x**2)
def j(x):
    return (1((1/2)*x)-((1/8)*(x**2))+((1/8)*(x**3)))
```

```python
#Valores de eje x que toma en la grafica.
x=np.linspace(-1,10,100)
```

```python
# Grafica de funciones.
plt.plot(x, f(x),label="$sqrt(x+1)$")
plt.plot(x, g(x),label="$1+1/2洧논$")
#plt.plot(x, h(x),label="$1+1/2洧논-1/8x^{2}$")
#plt.plot(x, j(x),label="$1+1/2洧논-1/8洧논^{2}+1/8x^{3}$")
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-1, 10)
plt.ylim(-1, 5)
#mostrarlo
plt.grid()
plt.title("Grafica de $f(x)=sqrt(x+1)$")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.legend()
plt.show()
```
