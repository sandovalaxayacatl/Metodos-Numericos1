```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
#defenir la funcion g(x) que usaremos en el punto fijo
def f(x):
  #aqui defines tu funcion de iteracion f(x)
  return x**3-x-1
  #aqui defines la derivada de tu funcion f(x)
def df(x):
  return 3*x**2-1
```

```python
#parametros de entrada
p0 = 1 #valor inicial de p0
tol = 0.0001 #tolerancia
nmax = 100 #numero maximo de iteraciones
```

```python
#inicializar variables
error = 100  #error de la primera iteracion
niter = 0    #numero de iteraciones el cual ira incrementando
p = p0       #valor de p inicial
```

```python
print("#iter\t\t p \t\t\t f(p) \t\t\t error")  #celdas de la tabla
print("{0}\t\t {1:6.6f} \t\t {2:6.6f} \t\t {3:6.6f}".format(niter,p,f(p),error)) #primera linea de valores de los valores inicializados

# Método de Newton-Raphson
while error > tol and niter < nmax:
    niter += 1
    p = p0 - (f(p0) / df(p0))
    error = abs(p - p0)
    p0 = p
    #imprimir los resultados de la iteracion
    print("{0}\t\t {1:6.6f} \t\t {2:6.6f} \t\t {3:6.6f}".format(niter,p,f(p),error))

#imprimir el resultado final
print("La aproximacion de la raiz es:{0:6.7f} despues de {1} iteraciones".format(p,niter))
```
