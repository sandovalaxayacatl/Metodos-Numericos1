```python
import numpy as np
```

```python
# Datos de ejemplo
#xnum = [1.0, 1.3, 1.6, 1.9, 2.2]  #Lista de los valores x0, x1, ..., xn.
#ynum = [0.7651977, 0.6200860, 0.4554022, 0.2818186, 0.1103623]  # Lista de los valores f(x0), f(x1), ..., f(xn).

h = int(input("¿Cuántos numeros deseas ingresar?: "))
# Inicializar listas para los valores de x y f(x)
xnum = []
ynum = []
for i in range(h):
    xv = float(input(f"Ingrese el valor de x{i+1}: "))  #valores x0, x1, ..., xn.
    yv = float(input(f"Ingrese el valor de f(x) en x{i+1}: "))  #valores f(x0), f(x1), ..., f(xn).
    xnum.append(xv)  #lista de numeros x
    ynum.append(yv)  #lista de f(x)

x = 1.5  # Punto donde se evalúa el polinomio interpolante
```

```python
def metodo_de_neville(x, x_values, ynum):

    n = len(xnum)
    # Crear la tabla Q inicializada con ceros
    Q = np.zeros((n, n))

    # Rellenar la primera columna con los valores de y_values
    for i in range(n):
        Q[i, 0] = ynum[i]

    # Aplicar el método de Neville
    for i in range(1, n):
        for j in range(1, i + 1):
            Q[i, j] = ((x - xnum[i - j]) * Q[i, j - 1] - (x - xnum[i]) * Q[i - 1, j - 1]) / (xnum[i] - xnum[i - j])

    return Q, Q[n - 1, n - 1]
```

```python
# Calcular el resultado
tabla_Q, aproximacion = metodo_de_neville(x, xnum, ynum)
```

```python
# Mostrar resultados
print("Tabla Q:")
print(tabla_Q)
print(f"\nEl valor aproximado de f({x}) es: {aproximacion}")
```
