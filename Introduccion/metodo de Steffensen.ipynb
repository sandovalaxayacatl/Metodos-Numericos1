```python
#defenir la funcion g(x) que usaremos en el punto fijo
def f(x):
  #en este caso como utilizamos funciones trigonometricas, necesiamos inicializar la viariable x como:
  #x = math.pi / 3  # 60 grados en radianes
  #aqui defines tu funcion de iteracion g(x)
  return cos(x) -2*x +3
  #return x**2 - 2
  #return x**5 - x**4 +2*x**3 -3*x**2 +x -4
```

```python
#parametros de entrada
p0 = 1 #valor inicial de p (NECESITAMOS DAR UN PUNTO TAL QUE f(a) = a)
tol = 0.00001 #tolerancia
nmax = 100 #numero maximo de iteraciones
```

```python
#inicializar variables
error = 100  #valor inicial del error
niter = 0    #valor inicial de las iteraciones
p=p0
```

```python
print("#iter\t p \t\t f(p) \t\t error")                                     #primera fila de la tabla indicando su estructura
print("{0}\t {1:6.6f} \t {2:6.6f} \t {3:6.6f}".format(niter,p,f(p),error))  #segunda fila de la tabla con los valores de inicio

#ciclo iterativo del metodo de punto fijo
while error > tol and niter < nmax:
    niter += 1  # Incrementando el contador de iteraciones
    fp = f(p) # Calcula f(p)
    gp = p + fp  # Calcula g(p0) = p0 + f(p0)
    ggp = gp + f(gp)  # Calcula g(g(p)) = f(g(p)) + g(p)

    # Calcula el numerador y denominador de la fórmula de Steffensen
    numerador = (gp - p)**2
    denominador = ggp -2*(gp) + p0

    # Nueva aproximación
    p0 = p + -(numerador / denominador)
    # Calcula el error como la diferencia entre las iteraciones
    error = abs(p0 - p)
    # Actualiza p0 para la siguiente iteración
    p = p0

    #imprimir los resultados de la iteracion
    print("{0}\t {1:6.6f} \t {2:6.6f} \t {3:6.6f}".format(niter,p,f(p),error))

#imprimir el resultado final
print("La aproximacion de la raiz es:{0:6.7f} despues de {1} iteraciones".format(p,niter))
```
