```python
#PROGRAMA PUNTO FIJO - Sandoval Hernandez Irving Axayacatl - 3MM1
from math import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
#defenir la funcion g(x) que usaremos en el punto fijo
def g(x):
  #aqui defines tu funcion de iteracion g(x)
  return (x+1)**(1/3)
  #return ((10)/(4+x))**(1/2)
```

```python
#parametros de entrada
p0 = 1 #valor inicial de p
tol = 0.0000001 #tolerancia
nmax = 100 #numero maximo de iteraciones
```

```python
#inicializar variables
error = 100
niter = 0
p = p0
```

```python
print("#iter\t\ p \t\t g(p) \t\t error")
print("{0}\t\t {1:6.6f} \t {2:6.6f} \t {3:6.6f}".format(niter,p,g(p),error))

#ciclo iterativo del metodo de punto fijo
while error > tol and niter < nmax:
  niter += 1  #incrementando el contador de iteraciones
  p_ant = p   #guardando el valor anterior de p
  p = g(p)    #calcular nuevo valor de p apartir de g(p)
  error = abs(p-p_ant) #calcular el valor como la diferencia de iteraciones

  #imprimir los resultados de la iteracion
  print("{0}\t\t {1:6.6f} \t {2:6.6f} \t {3:6.6f}".format(niter,p,g(p),error))

#imprimir el resultado final
print("La aproximacion de la raiz es:{0:6.7f} despues de {1} iteraciones".format(p,niter))
```
