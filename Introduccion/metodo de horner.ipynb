```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Función que evalúa el polinomio P(x) utilizando el método de Horner
def P(x, coef):
    n = len(coef)
    y = coef[0]  # Inicializamos y con el coeficiente principal (an)

    for i in range(1, n):
        y = x * y + coef[i]  # Método de Horner para evaluar P(x)
    return y
```

```python
# Función que evalúa la derivada del polinomio Q(x) utilizando el método de Horner
def Q(x, coef):
    n = len(coef)
    z = coef[0] * (n-1)  # Inicializamos z con el coeficiente principal multiplicado por el grado del polinomio

    for i in range(1, n-1):  # La derivada tiene un grado menor
        z = x * z + coef[i] * (n-1-i)  # Método de Horner para evaluar Q(x)

    return z
```

```python
# Coeficientes del polinomio P(x) = 2x^4 - 3x^2 + 3x - 4
coef = [2, 0, -3, 3, -4]  # Coeficientes en orden descendente de grado

# Parámetros de entrada
x0 = -2  # Valor inicial de x0
```

```python
# Evaluamos el polinomio P(x0) y su derivada Q(x0)
y = P(x0, coef)
z = Q(x0, coef)

#obtenemos la aproximacion a la raiz
x = x0 - y/z
```

```python
# Imprimimos los resultados
print(f"Evaluación de P(x) en x0 = {x0}: y = {y}")
print(f"Evaluación de la derivada Q(x) en x0 = {x0}: z = {z}")
print(f"Aproximación a la raíz: x = {x}")
```
