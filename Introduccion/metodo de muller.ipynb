```python
#Metodo de Muller - Sandoval Hernandez Irving Axayacatl - 3MM1
from math import *
import numpy as np
```

```python
def f(x):
  return x**3 - 13*x - 12
```

```python
#Datos de entrada:
p0=4.5
p1=5.5
p2=5
tol=0.00001 #tolerancia
nmax = 100 #numero maximo de iteraciones
```

```python
#Determine h1 = p1 − p0;
h1=p1-p0                         #h1 = p1 − p0
h2=p2-p1                         #h2 = p2 − p1
δ1 = ( f (p1) - f (p0))/h1       #δ1 = ( f (p1) − f (p0))/h1
δ2 = ( f (p2) - f (p1))/h2       #δ2 = ( f (p2) − f (p1))/h2
d = (δ2 - δ1)/(h2 + h1)          #d = (δ2 − δ1)/(h2 + h1)
i=3                              #i = 3
error = 100  #valor inicial del error
```

```python
print("#iter\t p0 \t\t f(p0)\t\t p1 \t\t f(p1)\t\t p2 \t\t f(p3) \t\t ")                          #primera fila de la tabla indicando su estructura
print("{0}\t {1:6.6f} \t {2:6.6f} \t {3:6.6f} \t {4:6.6f} \t {5:6.6f} \t {6:6.6f}".format(i,p0,f(p0),p1,f(p1),p2,f(p2)))  #segunda fila de la tabla con los valores de inicio

#ciclo iterativo del metodo de Muller
while i < nmax:
    b = δ2 + h2*d
    D = (b**2 -4*(f(p2)*d))**(1/2)
    if abs(b-D) < abs(b+D):
        E = b + D
    else:
        E = b - D

    h = (-2*f(p2))/E
    p = p2 + h

    p0 = p1
    p1 = p2
    p2 = p
    h1 = p1 - p0
    h2 = p2 - p1
    δ1 = ( f (p1) - f (p0))/h1
    δ2 = ( f (p2) - f (p1))/h2
    d = (δ2 - δ1)/(h2 + h1)
    i = i + 1
    aprox=abs(h)
    if aprox < tol:
      break

    #imprimir los resultados de la iteracion
    print("{0}\t {1:6.6f} \t {2:6.6f} \t {3:6.6f} \t {4:6.6f} \t {5:6.6f} \t {6:6.6f}".format(i,p0,f(p0),p1,f(p1),p2,f(p2)))

#imprimir el resultado final
print("La aproximacion de la raiz es:{0:6.7f} despues de {1} iteraciones".format(p,i))
```
