```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import BarycentricInterpolator
```

```python
#Obtener los datos del usuario
n = int(input("¿Cuántos elementos deseas ingresar en tu tabla? "))

# Inicializar listas para los valores de x y f(x)
x = []
f = []
```

```python
# Obtener los valores de x y f(x) del usuario
for i in range(n):
    xv = float(input(f"Ingrese el valor de x{i+1}: "))
    fv = float(input(f"Ingrese el valor de f(x) en x{i+1}: "))
    x.append(xv)
    f.append(fv)
```

```python
#Graficar los puntos y el polinomio interpolante
plt.figure(figsize=(10, 6))
plt.scatter(x, f, color="red", label="Puntos dados")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfica de datos")
plt.legend()
plt.grid(True)
#plt.grid(color="black", linestyle="--")
plt.show()
```

```python
# Convertir las listas a arrays de numpy para facilitar el cálculo
x = np.array(x)
f = np.array(f)
```

```python
# Crear el polinomio interpolante de Newton
polinomio_interpolante = BarycentricInterpolator(x, f)
```

```python
# Paso 3: Preparar los datos para la gráfica
# Crear un rango de valores x para graficar el polinomio interpolante
x_plot = np.linspace(min(x), max(x), 100)
f_plot = polinomio_interpolante(x_plot)
```

```python
#Graficar los puntos y el polinomio interpolante
plt.figure(figsize=(10, 6))
plt.plot(x_plot, f_plot, label="Polinomio Interpolante de Newton", color="blue")
plt.scatter(x, f, color="red", label="Puntos dados")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Interpolación de Newton")
plt.legend()
plt.grid(True)
#plt.grid(color="black", linestyle="--")
plt.show()
```
